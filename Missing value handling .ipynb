```python
#droped the 1st,5th,10th,11th row
df = df.drop(index=[1,5,10,11])
print(df)
```

```python
# Fill NaN Null values with the mode
import numpy as np
df = df.replace(['Null', 'unknown', 0, 'n/a'], np.nan)
df['Marital status'].fillna(df['Marital status'].mode()[0], inplace=True)

print(df)
```

```python
df = df.replace(['Null', 'unknown', 0, 'n/a'], np.nan)
df['Age'].fillna(df['Age'].mean(), inplace=True)

print(df)
```

```python
# Convert 'Income' column to numeric, coercing errors to NaN
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
# Now fill NaN values with the mean
df = df.replace(['Null', 'unknown', 0, 'n/a'], np.nan)
df['Income'].fillna(df['Income'].mean(), inplace=True)

print(df)
```

```python
df.to_csv('processed_data.csv', index=False)
```
